```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
import matplotlib.pyplot as plt; plt.show()
matplotlib.rcParams.update({'font.size': 15})
```

```python
def f(x, w, sig):
    return w*np.exp(-x**2/(2*sig**2))
```

```python
xgrid = np.linspace(-8,8, 1000)
```

```python
plt.figure(figsize=(8,7))
plt.plot(xgrid, f(xgrid, 1, 1), lw=4)
plt.axvline(x=-5, c="red", lw=3)
plt.axvline(x=5, c="red", lw=3)
plt.xlabel("x-axis")
plt.text(-8.7, 0.5,r'$u(-5) = 3.7e-06$', size=12.5)
plt.text(5.1, 0.5, r'$u(5) = 3.7e-06$', size=12.5)
plt.grid()
plt.show()
```

```python
xgrid = np.linspace(-1,1, 1000)
plt.figure(figsize=(8,7))
plt.plot(xgrid, f(xgrid, 0.1, 0.1), lw=4)
plt.axvline(x=-0.5, c="red", lw=3)
plt.axvline(x=0.5, c="red", lw=3)
plt.xlabel("x-axis")
#plt.text(-8.7, 0.5,r'$u(-5) = 3.7e-06$', size=12.5)
#plt.text(5.1, 0.5, r'$u(5) = 3.7e-06$', size=12.5)
plt.grid()
plt.show()
```

```python
f(0.5, 0.1, 0.1)
```

Recordad que el dominio de nuestra imágen está acotado a [0,1]x[0,1] y que los valores de intensidad también están acotados a [0,1]. Luego este corresponde al peor caso: Un Gaussian component con peso $w_i=1$ y $\sigma_i = 1$. En la práctica los componentes de la mixture cumple $w_i \ll 1$ y $\sigma_i \ll 1$, y por lo tanto el error de aproximación a $5 \sigma_i$ es mucho menor.
