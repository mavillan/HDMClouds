```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
import matplotlib.pyplot as plt; plt.show()
matplotlib.rcParams.update({'font.size': 15})

save_path = "/Users/martin/Dropbox/Aplicaciones/ShareLaTeX/HDMClouds-thesis/figures/"
```

```python
def gf(x, w, mu, sig):
    return w*np.exp(-0.5*(x-mu)**2/(sig**2))
```

```python
xgrid = np.linspace(-4,4,250)
```

## KL approach for hierarchical clustering

```python
g1 = gf(xgrid, 0.09, -3, 0.35)
g2 = gf(xgrid, 0.25, -2.14285714, 0.5)
g3 = gf(xgrid, 0.4, -1.28571429, 0.25)
g4 = gf(xgrid, 0.23, -0.40857143, 0.42)
g5 = gf(xgrid, 0.1, 0.42857143, 0.6)
g6 = gf(xgrid, 0.05, 1.28571429, 0.35)
g7 = gf(xgrid, 0.15, 2.14285714, 0.31)
g8 = gf(xgrid, 0.24, 3, 0.49)
```

```python
plt.figure(figsize=(10,8))
plt.plot(xgrid, g1, lw=3, c="orange", label="cluster 1")
plt.plot(xgrid, g2, lw=3, c="red", label="cluster 2")
plt.plot(xgrid, g3, lw=3, c="green", label="cluster 3")
plt.plot(xgrid, g4, lw=3, c="gold", label="cluster 4")
plt.plot(xgrid, g5, lw=3, c="blue", label="cluster 5")
plt.plot(xgrid, g6, lw=3, c="c", label="cluster 6")
plt.plot(xgrid, g7, lw=3, c="hotpink", label="cluster 7")
plt.plot(xgrid, g8, lw=3, c="purple", label="cluster 8")

plt.tick_params(axis='both', which='major', labelsize=0)
plt.tick_params(axis='both', which='minor', labelsize=0)
plt.grid()
plt.plot(xgrid, g1+g2+g3+g4+g5+g6+g7+g8+0.025, lw=3, c="black")
plt.legend(loc=1)
plt.savefig(save_path+"HC-scheme-KL-01.eps", format='eps', dpi=50, bbox_inches='tight')
plt.show()
```
